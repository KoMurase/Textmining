```
import MeCab
def Mecab(text):
  mecab = MeCab.Tagger()
  result = ''

  text = text
  mecab.parse('')#文字列がGCされるのを防ぐ
  node = mecab.parseToNode(text)
  while node:
      #単語を取得
      word = node.surface
      #品詞を取得
      pos = node.feature.split(",")[1]
      #print('{0} , {1}'.format(word, pos))
      result += '{},{} \n'.format(word, pos)
      #次の単語に進める
      node = node.next
  return result
```

```
def use_Cabocha_write_result(data,file_):
  count = 0
  for sen in data.split('。'):
    

    result = ''
    c = CaboCha.Parser() 
    result = c.parseToString(sen)

    dir = '/content/drive/My Drive/TM/result_cabocha/'
    file_name = dir + "{}".format(file_)

    if os.path.exists(file_name):
      file = open(file_name, "a", encoding="utf_8")
      file.write(result)
      file.close() 
    else:
      file = open(file_name, "w", encoding="utf_8")
      file.write(result)
      file.close() 

    count+=1
    if count==30:
      break

  del file, data
  gc.collect() 
  print('{} is done.'.format(file_) )
```
